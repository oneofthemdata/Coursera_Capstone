```python
#Pandas/NumPy for data manipulation
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_list = pd.read_html(url)
```

```python
df_raw = df_list[0]
df_raw
```

#### Ignoring cells with a borough that is Not assigned.

```python
df_raw = df_raw[df_raw.Borough != 'Not assigned']
df_raw
```

#### Unlike mentioned in instruction there is no more than one neighborhood can exist in one postal code area.

```python
df_raw[df_raw.duplicated(subset=['Postal Code'])]
```

#### Counting "Not assigned" and "NaN" values in Neighbourhood and Borough columns

```python
selected_words_neigh=["Not assigned","NaN"]

df_raw.loc[df_raw['Neighbourhood'].isin(selected_words_neigh),'Neighbourhood'].value_counts().sum()
```

```python
selected_words_boro=["Not assigned", "NaN"]

df_raw.loc[df_raw['Borough'].isin(selected_words_boro),'Borough'].value_counts().sum()
```

#### Sorting values in Postal Code column

```python
df = df_raw[['Postal Code', 'Borough', 'Neighbourhood']]
df = df.sort_values(['Postal Code'])
df
```
